# GitHub Repository Preparation

Helper notebook used to download GitHub repositories for the main RAG project.

Output:

- source files for further processing

## Setup

Import libraries for GitHub access and repository cloning.

```python
import os
from dotenv import load_dotenv
from github import Github
from git import Repo
from pathlib import Path
```

```python
load_dotenv(override=True)
token = os.getenv("GITHUB_TOKEN")
```

## GitHub Connection and clone Repositories

Connect to GitHub and access user repositories, then clone them.

```python
g = Github(token)
user = g.get_user("JoannaWaligora")

ROOT = Path("..")
BASE_DIR = ROOT / "repos"

BASE_DIR.mkdir(exist_ok=True)

for repo in user.get_repos():
    repo_path = BASE_DIR / repo.name

    if not repo_path.exists():
        print(f"Cloning {repo.name}...")
        Repo.clone_from(repo.clone_url, str(repo_path))
    else:
        print(f"{repo.name} already exists")
```
